## Looping in dictionaries

```python
dict1={"name":"mrunal","age":22,"city":"pune"}
```

```python
for i,j in dict1.items():
    print("key is:",i,"value is :",j)
```

# i will have key 
# j will have value

```python
for i in dict1.keys():
    print(i)
```

```python
for j in dict1.values():
    print(j)
```
