## AND operator


```python
age=int(input("enter your age:"))
nationality=int(input("enter your nationality:"))

print(age)

print(nationality)

if (age>=18) and (nationality=="indian"):
   print("you are eligible for voting")

print("Program executed sucessfully")

```
