```python
departments = ["HR", "IT", "Finance", "Marketing"]
print("-- Using for loop --")
for dept in departments:
    print(dept)

print("-- Using enumerate --")
for idx, dept in enumerate(departments):
    print(idx, dept)
```

```python
grades = [55, 72, 48, 90, 65, 38, 80]
for idx, val in enumerate(grades):
    if val >= 50:
        print(f"Index {idx}: {val} -- Pass")
    else:
        print(f"Index {idx}: {val} -- Fail")
```

## Q20

```python
inventory = ["Pen", "Pencil", "Eraser", "Sharpener", "Notebook"]
print("First item:", inventory[0])
print("Last item:", inventory[-1])
print("Middle items:", inventory[1:4])
print("Every alternate:", inventory[::2])
for pos, prod in enumerate(inventory):
    print(pos, "-->", prod)
```
